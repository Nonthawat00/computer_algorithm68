```python
Fname = input("Fname is: ")
Lname = input("Lname is: ")
score = input("score is: ")
score = int(score)
if score >= 80:
    print(Fname,Lname,"A")
elif score >= 75:
    print(Fname,Lname,"B+")
elif score >= 70:
    print(Fname,Lname,"B")
elif score >= 65:
    print(Fname,Lname,"C+")
elif score >= 60:
    print(Fname,Lname,"C")
elif score >= 55:
    print(Fname,Lname,"D+")
elif score >= 50:
    print(Fname,Lname,"D")
else:
    print(Fname,Lname,"F")

```
